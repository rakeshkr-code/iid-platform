Key observation from the official docs:

**What is mentioned?**
- DocumentConverter → converts documents
- StandardPdfPipeline → processing pipeline
- Output → DoclingDocument (structured)

**What is NOT mentioned anywhere**
- ❌ No API to render page regions
- ❌ No API to export images from bbox
- ❌ No guarantee that pictures[i].image is populated
- ❌ No method like get_image() / render_region()

👉 That omission is intentional.

```
          Docling
            ↓
   structured layout (texts, pictures, bbox)
            ↓
     + external renderer
            ↓
   cropped images aligned with structure
```

1. Open PDF once (important for performance)

````python
import fitz  # PyMuPDF

pdf = fitz.open(file_path)
````

2. Function to crop any Docling picture

````python
def render_picture(picitem, pdf, zoom=2):
    # Get page (Docling is 1-based, PyMuPDF is 0-based)
    page_no = picitem['prov'][0].page_no - 1
    page = pdf[page_no]

    bbox = picitem['prov'][0].bbox

    # Create rectangle (Docling uses bottom-left origin → compatible)
    rect = fitz.Rect(bbox.l, bbox.b, bbox.r, bbox.t)

    # Increase resolution (important for RAG quality)
    mat = fitz.Matrix(zoom, zoom)

    pix = page.get_pixmap(matrix=mat, clip=rect)

    return pix
````

````python
def render_picture(picitem, pdf, zoom=2):
    page_no = picitem['prov'][0].page_no - 1
    page = pdf[page_no]

    bbox = picitem['prov'][0].bbox
    page_height = page.rect.height

    # Correct coordinate transform
    x0 = bbox.l
    x1 = bbox.r
    y0 = page_height - bbox.t
    y1 = page_height - bbox.b

    rect = fitz.Rect(x0, y0, x1, y1)

    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat, clip=rect)

    return pix
````

3. Save or view the image

````python
for picitem in docdict['pictures']:
    picitem = dict(picitem)

    if picitem['prov'][0].page_no == 5:
        pix = render_picture(picitem, pdf)

        out_path = f"{picitem['self_ref'].split('/')[-1]}.png"
        pix.save(out_path)

        print(f"Saved: {out_path}")
````
