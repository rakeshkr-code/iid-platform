## Source acquisition and file classification

- This layer code should be framework-agnostic and should be run it before Docling/LangChain.
- Docling is strong for downstream parsing because it supports many formats and preserves text, tables, pictures, hierarchy, layout boxes, and provenance in a unified document model.
- OCRmyPDF is a good OCR fallback because it adds searchable text layers to scanned PDFs and is tolerant of born-digital PDFs that do not need OCR.
- LangChain can then consume the cleaned outputs later through loaders such as DoclingLoader, Unstructured, JSONLoader, BSHTMLLoader, and CSVLoader.

- Now, let's observe all the text items from page no 3

```python
for textitem in docdict['texts']:
    textitem = dict(textitem)
    if textitem['prov'][0].page_no == 3:
        print(f"{textitem['self_ref']} -----------")
        pprint(textitem)
        print()
```

**Observations**
- Multicolumn and multiblocks of texts
- Among all text item blocks 'DocItemLabel.TEXT' are extracted correctly
- But 'DocItemLabel.CAPTION', 'DocItemLabel.PAGE_FOOTER' or 'DocItemLabel.SECTION_HEADER' are not extracted correctly
